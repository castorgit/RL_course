#### **Monte Carlo simulation Monty Hall Problem**

```python
import matplotlib.pyplot as plt
import random
```

```python
switch_win_probability = []
stay_win_probability = []

doors = ['car', 'goat', 'goat']
```

```python
def monte_carlo_simulation(iterations, verbose=True):
    
    win_switch_counter = 0
    win_stay_counter = 0
    
    for i in range (iterations):
        random.shuffle(doors)
        
        ini_choice = random.randrange(0,2)
        
        if doors[ini_choice] != 'car':               # initial choice is not car
            win_switch_counter += 1                  # switching is right (the host will switch to no car)
        else:
            win_stay_counter += 1                    # if is a car staying is the right choice
        
        # Probabilities
        win_by_switch = win_switch_counter / (i+1)
        win_by_stay   = win_stay_counter   / (i+1)
        
        # update the lists
        switch_win_probability.append(win_by_switch)
        stay_win_probability.append(win_by_stay)
        
        # Return final values
        if verbose:
            print(doors, win_by_switch, win_by_stay)
    return switch_win_probability[-1], stay_win_probability[-1]
```

```python
switch, stay = monte_carlo_simulation(100000, False)
```

```python
print(switch, stay)
```

```python
plt.figure(figsize=(15,5))
plt.plot(switch_win_probability, label= "switch")
plt.plot(stay_win_probability,  label= "stay")
plt.xlim(0,200)
plt.legend()
plt.xlabel('iterations')
plt.ylabel('win probabilities')
plt.show()
```
