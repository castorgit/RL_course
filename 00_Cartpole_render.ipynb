```python
import gymnasium as gym
import matplotlib.pyplot as plt
```

```python
# rgb-array
env = gym.make('CartPole-v1', render_mode="rgb_array")
env.reset()

for _ in range(1):
    frame = env.render()  # Get frame as an RGB array
    action = env.action_space.sample()
    env.step(action)
    
env.close()

plt.imshow(frame)
plt.show()
```

```python
# human
env = gym.make('CartPole-v1', render_mode="human")
env.reset()

for _ in range(1):
    frame = env.render()  # Get frame as an RGB array
    action = env.action_space.sample()
    env.step(action)
    env.render()    
```
