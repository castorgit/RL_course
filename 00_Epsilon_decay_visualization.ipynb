### **Visualization epsilon decay**

Using greedy selection of action we use epsilon to select exploration or exploitaition <br>
with this notebook you can compare how is the structure of different epsilon values and how fast<br>
they will go to the minimum.<br>
Each problem needs different slope.....and slope depends on the number of episodes to converge....

```python
import matplotlib.pyplot as plt
```

```python
epsilon = 1
NUMBER_EP = 1000
epsilon_decay1 = 0.9995
epsilon_decay5 = 0.999
epsilon_decay2 = 0.995
epsilon_decay3 = 0.99
epsilon_decay4 = 0.985

epsilon_min = 0.01

e1 = [max(epsilon_min, epsilon := epsilon * epsilon_decay1) for _ in range (NUMBER_EP)]
epsilon = 1
e2 = [max(epsilon_min, epsilon := epsilon * epsilon_decay2) for _ in range (NUMBER_EP)]
epsilon = 1
e3 = [max(epsilon_min, epsilon := epsilon * epsilon_decay3) for _ in range (NUMBER_EP)]
epsilon = 1
e4 = [max(epsilon_min, epsilon := epsilon * epsilon_decay4) for _ in range (NUMBER_EP)]
epsilon = 1
e5 = [max(epsilon_min, epsilon := epsilon * epsilon_decay5) for _ in range (NUMBER_EP)]

plt.plot(e1, label = str(epsilon_decay1))
plt.plot(e2, color='red',   label = str(epsilon_decay2))
plt.plot(e3, color='green', label = str(epsilon_decay3))
plt.plot(e4, color='orange', label = str(epsilon_decay4)) 
plt.plot(e5, color='yellow', label = str(epsilon_decay5)) 

# Add labels and title
plt.xlabel('steps')
plt.ylabel('epsilon value')
plt.legend()
plt.title('Line plot of epsilon values for '+str(NUMBER_EP)+' episodes')
plt.show()
```
